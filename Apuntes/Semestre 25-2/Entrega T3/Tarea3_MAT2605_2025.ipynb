a)  Para cada uno de los siguiente encuentre una matriz $A$ que satisface la condición

- El método de SOR converge para $A$ pero no converge el método de Gaus-Seidel.

Justifique su elección apropiadamente.

b) Sean las matrices
$$
A = \begin{bmatrix} a & c & 0 \\ c & a & c \\ 0 & c & a \end{bmatrix}, \quad
B = \begin{bmatrix} 0 & b & 0 \\ b & 0 & b \\ 0 & b & 0 \end{bmatrix},
$$
- Probar que $\lim_{n\to\infty} B^n = 0$, si y sólo su $|b|<\sqrt{2}/2$.
- Dar condiciones necesarias y suficientes sobre $a,c$ reales, para la convergencia de los métodos de Jacobi y Gauss-Seidel aplicados a la resolución de $Ax = v$.


<div class="alert alert-block alert-success"> <b> Solución. </b> </div>

a) Sabemos por teorema que para cualquier método iterativo si se tiene que $\rho(T)<1$ el método converge, también se sabe que para matrices estrictamente diagonal dominantes Jacobi y Gauss-Seidel convergen, mientras que para matrices definidas positivas Gauss-Seidel y SOR tienen posibilidad de converger. Hay que igual tener en cuenta la elección del parámetro $\omega$ para el método SOR.

- $A = \begin{bmatrix} 2 & 1  \\ 1 & 2  \end{bmatrix}$ es una matriz que converge para Jacobi ya que es diagonal dominante, luego para que no converga SOR basta con elegir $\omega$ fuera del rango $(0,2)$.

- $A = \begin{bmatrix} 2 & -5 & 1 \\ 1 & 4 & 6 \\ -1 & 4 & 2 \end{bmatrix}$. Elegí una matriz $3\times 3$ que cumpliera con: 1. Tiene que ser una matriz definida positiva para que SOR tenga posibilidad de converger. 2. No tiene que ser estrictamente diagonal dominante para que así no converja para Jacobi ni Gaus-Seidel. La matriz elegida tiene subdeterminantes:
$$ \Delta_1 = 2 \quad \Delta_2 = 13 \quad \Delta_3 = 16 \quad :\Delta_i >0 $$ También, no es estrictamente diagonal dominante ya que: $$ \text{Fila}_1: 2<|-5| \quad \text{Fila}_2: 4<6 \quad \text{Fila}_3: 2<4 \quad $$

Se codifica a continuación el cálculo de las matrices de iteración de los métodos trabajados, se puede observar que para $\omega=0.4$ SOR converge ya que $\rho(T_{SOR}) < 1 $, mientras que los otros radios espectrales son mayores a 1.

```python
def probar_matriz(A, omega = 1.0, method = 'Jacobi'):
    D = np.diag(np.diag(A))
    L = np.tril(A, -1)  
    U = np.triu(A, 1)   

    if method == 'Jacobi':
        # T_J = -D^-1 (L+U)
        TJ = -np.linalg.inv(D) @ (L + U)
        rho = max(abs(np.linalg.eigvals(TJ)))
        return rho
    
    elif method == 'Gauss-Seidel':
        # T_GS = -(D + L)^-1 U
        TGS = -np.linalg.inv(D + L) @ U
        rho = max(abs(np.linalg.eigvals(TGS)))
        return rho
    
    elif method == 'SOR':
        # SOR = (omega*L + D)^-1 * ((1-omega)*D - omega*U)
        T_SOR = np.linalg.inv(omega*L + D) @ ((1-omega)*D - omega*U)
        rho = max(abs(np.linalg.eigvals(T_SOR)))
        return rho

A = np.array([[2, 1],
              [1, 2]])

print("rho Jacobi:", probar_matriz(A, method='Jacobi'))
print("rho Gauss-Seidel:", probar_matriz(A, method='Gauss-Seidel'))
print("rho SOR:", probar_matriz(A, omega=2, method='SOR'))

print(' ')

A = np.array([[2, -5, 1],
              [1, 4, 6],
              [-1, 4, 2]])

print("rho Jacobi:", probar_matriz(A, method='Jacobi'))
print("rho Gauss-Seidel:", probar_matriz(A, method='Gauss-Seidel'))
print("rho SOR:", probar_matriz(A, omega=0.4, method='SOR'))
```

b.i) Nos interesa mostrar que la matriz $B$ es nilpotente, del curso de EDO recordamos que una matriz es nilpotente si todos sus valores propios son 0. Esto es porque la matriz B tiene diagonalización $B = PDP^{-1}$, entonces: $B^k = PD^kP^{-1}$ y elevar la matriz diagonal $D$ a la k es simplemente elevar los valores propios a la k. Concluimos que como:
$$ D^k = \begin{bmatrix} \lambda^k & 0 & 0  \\ 0 & \lambda^k & 0 \\  0 & 0 &\lambda^k  \end{bmatrix} \to  0 \quad ; \quad \lambda^k \to 0 \iff |\lambda| < 1 $$

Calculando los valores propios de B se obtiene el polinomio característico:$$ -\lambda^3 +2b^2\lambda = 0 \leftrightarrow \lambda_1 = 0 , \lambda_{2,3}=\pm \sqrt{2b^2} $$ 
Luego $$ |\lambda_{2,3}| < 1 \ , \ |\sqrt{2b^2}| <1 \ , \ |b|<\frac{\sqrt{2}}{2}$$

b.ii) Sabemos por teorema que los métodos de Jacobi y Gaus-Seidel convergen para matrices estrictamente diagonal dominantes. Luego un criterio suficiente para la convergencia es que: $$|a| > 2|b|$$ Calculando los radios espectrales de las matrices de iteración para $A$: $$\rho(A_{J}) = \sqrt 2 |\frac{c}{a}| \quad , \quad \rho(A_{GS}) = 2\frac{c^2}{a^2}$$
Obtenemos las condiciones: $|\frac{c}{a}| < \frac{\sqrt 2}{2} \quad , \quad c^2<\frac{a^2}{2} $.

<div class="alert alert-block alert-info"> <b> Problema 2. Métodos iterativos (práctica)</b> </div>

Sea $A\in \mathbb R^{n\times n}$ dada por
\begin{equation*}
A = 
\begin{pmatrix}
2 & -1 & 0 & \ldots & 0 \\
-1& 2 & -1 &\ddots & \vdots\\
0 & \ddots & \ddots & \ddots & 0\\
\vdots & \ddots & -1 &2& -1 \\
0 & \ldots & 0 & -1 & 2
\end{pmatrix}
\end{equation*}
y la matriz de Hilbert  $H_n\in \mathbb R^{n\times n}$ definida por 
$$
(H_{n})_{i,j} = \frac{1}{i+j-1}, \qquad \text{para } i,j=1,...,n. 
$$

Programe los métodos iterativos de Jacobi, Gauss-Seidel y SOR y calcule para $n=10$ la solución de los sistemas lineales con $b = (1,...,1)^T\in \mathbb R^{10}$, e iniciando con $x^{(0)} = b$.
Dada la tolerancia $10^{-12}$ registre en un tabla el número de iteraciones de cada uno de los métodos. Comente sus resultados.

<div class="alert alert-block alert-success"> <b> Solución. </b> </div>

```python
def jacobi(A, b, x0, max_iteracion = 100, tol = 1e-12):
    x = x0.copy()
    D = np.diag(np.diag(A))
    L = np.tril(A, -1)
    U = np.triu(A, 1)
    T_J = -np.linalg.inv(D) @ (L+U)
    C_J = np.linalg.inv(D).dot(b)
    errores = []

    for i in range(max_iteracion):
        error = np.linalg.norm(b - np.dot(A, x))
        errores.append(error)

        if error < tol:
            print(f"Convergencia alcanzada en {i+1} iteraciones.")
            return x, errores
        x = T_J.dot(x) + C_J
    
    print(f'   {i+1}   |   {error}   ')
    print('Máximo de iteraciones alcanzado\n')
    return x, errores

def gauss_seidel(A, b, x0, max_iteracion = 100, tol = 1e-12):
    x = x0.copy()
    n = x.size
    D = np.diag(np.diag(A))
    L = np.tril(A)
    U = np.triu(A)
    T_GS = -np.linalg.inv(D + L) @ U
    C_GS = np.linalg.inv(D+L) @ (b)
    errores = []

    for i in range(max_iteracion):
        error = np.linalg.norm(b - np.dot(A, x))
        errores.append(error)

        if error < tol:
            print(f"Convergencia alcanzada en {i+1} iteraciones.")
            return x, errores
        x = T_GS.dot(x) + C_GS

    print(f'   {i+1}   |   {error}   ')
    print('Máximo de iteraciones alcanzado\n')
    return x, errores

def sor(A, b, x0, w = 1.5, max_iteracion = 100, tol = 1e-12):
    x = x0.copy()
    n = x.size
    D = np.diag(np.diag(A))
    L = np.tril(A)
    U = np.triu(A)
    T_SOR = np.linalg.inv(w*L+D) @ ((1-w)*D-w*U)
    C_SOR = np.linalg.solve(D+w*L,w*b)
    errores = []

    for i in range(max_iteracion):
        error = np.linalg.norm(b - np.dot(A,x))
        errores.append(error)
        if error < tol:
            print(f"Convergencia alcanzada en {i+1} iteraciones.")
            return x, errores
        x = T_SOR.dot(x) + C_SOR

    print(f'   {i+1}   |   {error}   ')
    print('Máximo de iteraciones alcanzado\n')
    return x, errores
```

```python
def make_hilbert(n):
    hilbert = np.ones((n,n))
    for i in range(1, n):
        for j in range(1, n):
            a = 1 / (i+j-1)
            hilbert[i][j] = a
    return hilbert

def make_tridiagonal(n):
    trid = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            if i == j:
                trid[i][j] = 2
            elif j == (i+1) or j == (i-1):
                trid[i][j] = -1
    return trid

n = 10
x0 = np.ones(n)
H = make_hilbert(n)
A = make_tridiagonal(n)

# Jacobi
sol_jacobia, eja = jacobi(A, x0, x0)
sol_jacobih = jacobi(H, x0, x0)
# Gauss-Seidel
sol_seidela, ejg = gauss_seidel(A, x0, x0)
sol_seidelh = gauss_seidel(H, x0, x0)
# SOR
sol_sora = sor(A, x0, x0)
sol_sorh = sor(H, x0, x0)
```

Con estos resultados se puede apreciar que la matriz tridiagonal está mucho mejor condicionada que la matriz de Hilbert, los errores en la última se disparan, llegando al orden de $10^{73}$ en el método de Gauss-Seidel (que fue el menor). <div class="alert alert-block alert-info"> <b> Problema 3. Iteracion de potencia, inversa y cociente de Rayleigh (teorico)</b> </div>

Suponga que una matriz diagonalizable que tiene un par $\lambda_1, \lambda_2$ de valores propios dominantes con signo opuesto, esto es, $\lambda_1=-\lambda_2$, y $|\lambda_1|>|\lambda_i|$, para $i>2$. 
Explique como se puede obtener una aproximación usando la iteración de potencia a los dos vectores propios correspondientes.

<div class="alert alert-block alert-success"> <b> Solución. </b> </div>

Lo que hacemos en el método de la potencia es escribir un vector x en función de vectores propios, luego al multiplicar varias veces por la matriz $A$ queda lo siguiente: $$ Ax = \sum\limits^n_{j=1}\beta_j Av^{(j)} = \sum\limits^n_{j=1}\beta_j \lambda_jv^{(j)} $$
$$ A^kx = \sum\limits^n_{j=1}\beta_j \lambda_j^k v^{(j)}$$

Normalmente se factoriza por $\lambda_1^k$ (dominante), entonces todos los términos $(\frac{\lambda_i}{\lambda_1})^k \to 0$. Sin embargo En este caso al hacer este paso quedaría:
$$ A^kx = \lambda_i^k\sum\limits^n_{j=1}\beta_j (\lambda_j/\lambda_1)^k v^{(j)} \sim \lambda_1^k\beta_1v^{(1)}\pm\lambda_1^k\beta_2v^{(2)}$$
Esto ya que $(\lambda_2/\lambda_1)^k = -1^k$. 

La iteración inversa es un método muy parecido el método al de la potencia ya que también se multiplica una matriz por un vector de forma iterada, sin embargo, se multiplica por $A^{-1}$. Si tomamos la matriz $A^{-1} = (A-\mu I)$, entonces el método converge a los valores propios $(\lambda_i -\mu)^{-1}$ con $\lambda_i$ los valores propios de A. Si nos fijamos, el valor propio dominante del método inverso es el menor valor propio de $A$ desplazado $\mu$. Luego, si queremos converger a $\lambda_2$ se puede usar $\mu$ negativo o simplemente un valor cercano a $\lambda_2$. 

<div class="alert alert-block alert-info"> <b> Problema 4. Iteracion de potencia, inversa y cuociente de Rayleigh (practico)</b> </div>

Considere la matriz 
\begin{equation}
A(\alpha) = 
\begin{pmatrix}
\alpha & 2 & 3 & 13\\
5 & 11 & 10 & 8 \\
9 & 7 & 6 & 12 \\
4 & 14 & 15 & 1
\end{pmatrix}
\end{equation}
para $\alpha \in \mathbb R$. Calcule, para $\alpha = 30$ y $\alpha = -30$, los valores propios con  mayor y menor modulo utilizando:

- El método de la potencia con vector inicial $(1,1,1,1)^T$ y la iteración inversa con $\mu$ el centro del círculo de Gershgorin correspondiente.
- El método de la potencia con cociente de Rayleigh.

Registre el número de iteraciones en ambos casos y compárelos. Comente sus resultados. 

<div class="alert alert-block alert-success"> <b> Solución. </b> </div>

Para calcular el valor propio dominante de $A$, se toma el $\mu =$ centro del círculo de Gershgorin con mayor valor absoluto, así el método converge a $\frac{1}{\lambda_n}$, $\lambda_n$ siendo el dominante. Si utilizamos el $\mu$ con menor valor absoluto, el método converge a el menor valor propio de $A$, se puede simplemente utilizar también $\mu = 0$.

```python
def discos_de_Gershgorin(A):
    n = A[0].size
    discos = dict() # centro: radio
    for i in range(n):
        r = 0
        for j in range(n):
            if i != j:
                r += abs(A[i][j])
        centro = A[i][i]
        discos[centro] = r
    return discos

def iteracion_inversa(A, x0, calcular, max_iter = 100, tol = 1e-12):
    x = x0.copy()
    discos = discos_de_Gershgorin(A)
    # Aproximaciones de los valores propios mayor y menor:
    centros = np.array(list(discos.keys())) ; radios = np.array(list(discos.values()))
    mu = np.max(centros + radios) ; nu = np.min(centros - radios)
    p = np.argmax(np.abs(x))
    lambdas = [] ; xlist = []

    # Iteración 
    for i in range(max_iter):
        if calcular == 'max':
            y = np.linalg.solve(A - mu * np.eye(A.shape[0]), x)       
            x = y/np.linalg.norm(y)
            delta = y[p] 
            lam = mu + 1.0/delta
            lambdas.append(lam)

        elif calcular == 'min':
            y = np.linalg.solve(A - nu * np.eye(A.shape[0]), x)
            x = y/np.linalg.norm(y)
            delta = y[p] 
            lam = nu + 1.0/delta
            lambdas.append(lam)

        if i > 0 and abs(lambdas[-1] - lambdas[-2]) < tol:
            print('--------------------------------')
            print('Convergencia en iteración:', i+1)
            print('Valor estimado:', lam)
            print('--------------------------------')
            return lam, xlist, i+1

        xlist.append(x)
        p = np.argmax(np.abs(y)) 
        x = y/y[p]

    print('Máximo de iteración alcanzado ...')
    print(f'Valor propio estimado en iteración {i+1}:\n {lam}')

    return lam, xlist, max_iter
```

```python
def iteracion_rayleigh(A, x0, max_iter = 100, tol = 1e-12):
    x = x0.copy()
    # r = x.dot(A.dot(x))/(x.dot(x)) Cociente de Rayleigh
    lambdas=[]; xlist=[]

    for i in range(max_iter):
        y = A.dot(x)
        x = y / np.linalg.norm(y, ord = 2)
        lam = x.dot(A.dot(x))/(x.dot(x))
        lambdas.append(lam) ; xlist.append(x)
        
        if i > 0 and abs(lambdas[-1] - lambdas[-2]) < tol:
            print('--------------------------------')
            print('Convergencia en iteración:', i+1)
            print('Valor estimado:', lam)
            print('--------------------------------')
            return lambdas[-1], xlist, i+1

    print('Máximo de iteración alcanzado ...')
    print(f'Valor propio estimado en iteración {i+1}:\n {lambdas[-1]}')

    return lambdas[-1] , xlist, max_iter
```

```python
x0 = np.ones(4)
A1 = np.array([[30 , 2 , 3 , 13], 
               [5 , 11 , 10 , 8 ], 
               [9 , 7 , 6 , 12],
               [4 , 14 , 15 , 1]])
A2 = np.array([[-30 , 2 , 3 , 13], 
               [5 , 11 , 10 , 8 ], 
               [9 , 7 , 6 , 12],
               [4 , 14 , 15 , 1]])

r1 = iteracion_inversa(A1, x0, 'max')
print(' ')

r2 = iteracion_inversa(A1, x0, 'min')
print(' ')

r3 = iteracion_inversa(A2, x0, 'max')
print(' ')

r4 = iteracion_inversa(A2, x0, 'min')
print(' ')
```

A simple vista igualmente parece ser que ambos funcionan, eso sí, comparativamente me quedo con el método de la potencia inversa ya que puede calcular otros valores propios que no sean el dominante de la matriz. Esto sumado a que converge más rápido que el de la Potencia con Rayleigh. Esto muestra que para $\mu \approx \lambda$, el método de potencia inverso puede llegar a ser más rápido que el de la potencia con coeficiente de Rayleigh. 

Finalmente, sobre las matrices: $A_1$ con $\alpha = 30$ convergió comparativamente más lento que $A_2$ con $\alpha = -30$ en el método de iteración inversa. Esto también tiene reespaldo de la teoría ya que como mencioné anteriormente, el método inverso converge al autovalor mayor, luego la razón de convergencia del método es de $$ \frac{\lambda_j - \mu}{\lambda_i - \mu} $$

Luego como $\mu \approx \lambda_j$, la fracción es muy pequeña, por lo que el método converge rápido. Mientras mejor sea la aproximación, mejor la convergencia.
