Se tiene: $\frac{| f(x_1,x_2) - \tilde{f} (x_1,x_2) | }{ | f(x_1,x_2) |}$, reemplazando con las definiciones obtenemos: $\frac{| (x_1 + x_2) - fl (x_1+x_1\epsilon + x_2+x_2\epsilon) | }{ | (x_1+x_2) |}$. Sea $x_1 > x_2$, entonces tenemos que lo anterior es a lo más:
$$
\leq \frac{| (x_1 + x_1) - fl (x_1+x_1\epsilon + x_1+x_1\epsilon) | }{ | (x_1+x_1) |} = \frac{| (2x_1) - fl (2x_1+2x_1\epsilon) | }{ | (2x_1) |}
$$
Ahora, sea $2x_1 = x$, luego:
$$
\leq \frac{| x - fl (x+x\epsilon) | }{ | x |} 
$$
Por definición de punto flotante:
$
\frac{| x - fl (x+x\epsilon) | }{ | x |} = \frac{| x - (x+x\epsilon +x\epsilon +x\epsilon^2) | }{ | x |} =  \frac{| x - (x + 2x\epsilon + x\epsilon^2) | }{ | x |}
$
. Resolviendo el numerador obtenemos:
$$
\leq \frac{|  - 2x\epsilon - x\epsilon^2 | }{ | x |} = \frac{   2x\epsilon + x\epsilon^2  }{  x } = 2\epsilon +\epsilon^2 
$$

Ahora, se $g$ el problema de la resta de números reales $g(x_1,x_2) = x_1 - x_2 $.

Supongamos también que se tiene en la resta de números reales, una aproximación por punto flotante de la forma $fl(x) = x(1+\epsilon)$ y $\tilde{f}(x_1,x_2) = {fl}(x_1) \ominus {fl}(x_2)$. Si calculamos el error relativo que existe al restar dos numeros reales, se tiene:
$$ 
\frac{|f - \tilde{f}|}{|f|} = \frac{|(x_1-x_2) - (x_1(1+\epsilon) - x_2(1+\epsilon))(1+\epsilon)|}{|x_1 - x_2|} = \frac{|-(x_1 - x_2)\epsilon -x_1\epsilon(1+\epsilon)+x_2\epsilon(1+\epsilon)|}{|x_1-x_2|}
$$
Acotamos usando desigualdad triangular
$$
\leq |\epsilon| + \frac{|x_1|}{|x_1 -x_2|}|\epsilon|(1+\epsilon) + \frac{|x_2|}{|x_1 -x_2|}|\epsilon|(1+\epsilon)
$$

Acá notamos el problema de la resta, ya que en el cálculo de una cota, obtenemos una resta de dos números reales en los denominadores de dos fracciones. Luego, si se restan dos números similares i.e $x_1 \approx x_2 \implies |x_1 - x_2|$ es muy pequeño y el error se amplifica. Finalmente una cota para el problema de la resta de números reales, aproximando los epsilon que se consiguen al hacer los errores al representar $x_1,x_2$ como números de máquina es:
$$
\leq \epsilon + (\epsilon + \epsilon^2)(\frac{|x_1|+|x_2|}{|x_1-x_2|})
$$

<div class="alert alert-block alert-info"> <b> Problema 2. Evaluacion de integral</b> </div>

Consideremos el problema de evaluar las siguientes integrales en una maquina con precision finita
$$
E_n = \int_0^1 x^n e^{x-1}dx, \quad \text{para } n =0,1,2,...
$$
a) Deduzca una formula recursiva para el calculo de $E_n$, de la forma $E_n = a_n + b_n E_{n-1}$ (use integración por partes).

b) Demuestre que $E_n\to 0$ cuando $n\to \infty$.

c) De los siguientes algoritmos para el calculo de E_n ($n\geq 20$), decida cual prefiere y porque

- Calcular $E_0$ de forma analitica, y $E_n$ recursivamente por $E_k = a_k + b_k E_{k-1},\;k=1,2,...,n$.
- Hacer $E_m=0$, para $m>>n$ y  calcular recursivamente hacia atras $E_{k-1} = (E_k-a_k)/b_k,\; k =m, m-1,...,n+1$. 



<div class="alert alert-block alert-success"> <b> Solución. </b> </div>

a) Calculemos algunas integrales (por partes cuando se requiera):
$$ E_0 = \int_0^1 x^0 e^{x-1}dx = \int_0^1 e^{x-1}dx = \frac{1}{e}\int_0^1 e^x dx = \frac{1}{e}[e-1] = 1-\frac{1}{e} $$
$$ E_1 = \int_0^1 x^1 e^{x-1}dx = \int_0^1 xe^{x-1}dx = xe^{x-1} -\int e^{x-1}dx = xe^{x-1} - e^{x-1} = e^{x-1}(x-1)$$
Evaluando lo anterior se obtiene:
$$ E_1 = 0 - -\frac{1}{e} = \frac{1}{e} $$
$$ E_2 = \int_0^1 x^2 e^{x-1}dx = |^1_0 x^2 e^{x-1}-\int^1_0 e^{x-1}\cdot 2xdx = 1^2-2E_1 $$
$$ E_3 = \int_0^1 x^3 e^{x-1}dx =|^1_0 x^3 e^{x-1}-\int^1_0 e^{x-1}\cdot 3x^2dx =1^3 -3E_2 $$

Notamos la recurrencia cuando se integra por partes tomando $dv = e^{x-1}$, ya que obtenemos la integral del n anterior, luego obtenemos un cálculo recursivo de la forma:

$$ E_n = a_n +b_nE_{n-1} = 1-nE_{n-1} $$

b) Observemos que el integrando de $E_n$ es positivo por lo que podemos derivar una cota inferior para $E_n = 0$, ahora en el rango $x\in [0,1]$ se tiene:
- $0\leq x^n \leq 1$, para todo $n\in \mathbb{R}$.
- $0 < e^{x-1} \leq 1$, para $x\in [0,1]$.

Luego, podemos derivar una cota superior para $E_n$, reemplazando una de las expresiones por su valor máximo, reemplazamos $e^{x-1}$ por 1 para analizar que ocurre cuando $n\to \infty$:
$$
0 \leq E_n \leq \int_0^1 x^n \cdot 1 dx
$$
$$
0 \leq E_n \leq \mid_0^1 \frac{x^{n+1}}{n+1} = \frac{1}{n+1} / \lim\limits_{n\to \infty}
$$
$$
0 \leq \lim\limits_{n\to \infty}E_n \leq 0
$$

Por teorema del sandwich queda demostrado.

c) Analicemos la propagación del error. Supongamos que en el cálculo de la integral se comete un error $\Delta_n$, luego para cada itración se tiene:
$$\hat{E}_k = E_k +\Delta_k$$

- El primer algoritmo:
$$E_0 \ (analítico) \to E_1 \to \dots \to E_n \ (recursivamente)$$
En cada paso desde $E_0$, calculamos $E_k$ con su respectiva perturbación: $ \hat{E}_k = a_k +b_kE_{k-1} = a_k +b_k(E_{k-1}+\Delta_{k-1}) = E_k + b_k\Delta_{k-1} $.

Se observa entonces que el error se propaga  de la forma $\Delta_k = b_k\Delta{k-1}$, como calculamos $E_0$, el error final luevo de n pasos es:
$$ \Delta_n = (\prod_{k=1}^n b_k)\Delta_0 $$

Si $|b_k| > 1$, el error crece de manera exponencial.

- El segundo algoritmo:
$$E_m = 0 \ (\text{para } m >> n) \to E_{m-1} \to \dots \to E_n \ (recursivamente)$$

En cada paso se tienen errores de redondeo: $\hat{E}_{m-1} = (\hat{E}_m-a_m)/b_m = (\hat{E}_m - a_m)/b_m + \Delta_m $

El análisis muestra que el error se propaga como: $\Delta_{m-1} = \Delta_m / b_m + \epsilon_m$, luego el error final es:
$$ \Delta_n = (\frac{\Delta_{n+1}}{b_{n+1}} + \epsilon_{n+1}) \ , \ \Delta_{n+1} =  (\frac{\Delta_{n+2}}{b_{n+2}} + \epsilon_{n+2}) ... $$
Hasta llegar a $m, \text{con } E_m = 0$ , como es exacto $\Delta_m = 0$, luego:
$$ \Delta_{m-1} = \epsilon_{m} \ , \ \Delta_{m-2} = \frac{\Delta_{m-2}}{b_{m-2}} + \epsilon_{m-1} \dots $$

Se llega a:
$$ \Delta_n = \epsilon_{n+1} + \frac{\epsilon_{n+2}}{b_{n+1}} + \frac{\epsilon_{n+2}}{b_{n+1}b_{n+2}}+\dots + \frac{\epsilon_{m}}{b_{n+1}b_{n+2}\dots b_{m-1}} $$

Llegamos a la conclusión de que para $|b_k| > 1 $ los denominadores de las fracciones se hacen cada vez mayores, por lo que el error disminuye. 
- Se elige el algoritmo 1 para $ |b_k| < 1 $
- Se elige el algoritmo 2 para $ |b_k| > 1 $

A continuación describimos un algoritmo para calcular $\pi$.
Comenzar inicializando las variables $a, b, c, d, e$ del siguiente modo: 
$a = 0, b = 1, c = 1/\sqrt{2}, d= 1/4, e= 1$. Luego, iterar $n$ veces en el orden dado las siguientes formulas:
\begin{equation*}
a=b, \quad b= \frac{b+c}{2},\quad  c=\sqrt{ca},\quad d=d-e(b-a)^2,\quad e=2e.
\end{equation*}
Finalmente, el valor de $\pi$ puede estimarse por $f = b^{2}/d$, o por $g = (b+c)^{2}/(4d)$.

Escriba un pseudocódigo del algoritmo siguiendo el formato visto en clases. Luego programe el algoritmo para calcular los valores de $\pi$ estimados por $f$ y $g$ cuando $n = 1,2,..., 10$. ¿Que estimación converge más rápido? <div class="alert alert-block alert-success"> <b> Solución. </b> </div>

i) Pseudocódigo

<div class="alert alert-block alert-code">

**Algoritmo:** Estimar $\pi$ \
**Input:** $a,b,c,d,e \in \mathbb{R}\ , \ n\in \mathbb{N}$ \
**Output:** f \
$\text{para } i = 1,2,\dots,n:$ 

$\ a = b $  \
$\ b = \frac{b+c}{2}$\

$\ f = b^2 /d$\

</div>

<div class="alert alert-block alert-code">

**Algoritmo:** Estimar $\pi$.2 \
**Input:** $a,b,c,d,e \in \mathbb{R}\ , \ n\in \mathbb{N}$ \
**Output:** g \
$\text{para } i = 1,2,\dots,n:$ 

$\ a = b $  \
$\ b = \frac{b+c}{2}$\
$\ c = \sqrt{ca}$ \
$\ d = d-e(b-a)^2$ \
$\ e =2e$ \
$\ g = (b+c)^2 /4d$\

```python
# ii) códigos
import numpy as np

def estimar_pi(a: int, b: int, c: float, d: float, e: int, n: int, pi: float) -> float:
    print('| Estimación de pi 1 |\n')
    convergencia = False

    for i in range(n):
        a = b
        b = (b+c)/2
        c = np.sqrt(c*a)
        d = d - e*(b-a)**2
        e = 2*e
        f = float(b**2 /d)

        # Errores y demas
        relative_error = np.abs(pi - f) / np.abs(pi)
        print(f'El error relativo de la iteración {i+1} es de: {relative_error}')
        if relative_error < 1e-6 and not convergencia:
            print(f'*** Estimación converge en iteración {i+1} ***')
            convergencia = True
    print(f'\n*** VALOR FINAL DE ITERACIÓN: f = {f}')
    return f

def estimar_pi2(a: int, b: int, c: float, d: float, e: int, n: int, pi: float) -> float:
    print('| Estimación de pi 2 |\n')
    convergencia = False

    for i in range(n):
        a = b 
        b = (b+c)/2
        c = np.sqrt(c*a)
        d = d - e*(b-a)**2
        e = 2*e
        g = float((b+c)**2 / (4*d))

        # Errores
        relative_error = np.abs(pi - g) / np.abs(pi)
        print(f'El error relativo de la iteración {i+1} es de: {relative_error}')
        if relative_error < 1e-6 and not convergencia:
            print(f'*** Estimación converge en iteración {i+1} ***')
            convergencia = True
    print(f'\n*** VALOR FINAL DE ITERACIÓN: g = {g}')
    return g

# Iniciamos
pi = 3.14159265358979323846264338327950288 
a,b,c,d,e = 0,1,1/np.sqrt(2),1/4,1
n = 10
pi1 = estimar_pi(a,b,c,d,e,n,pi)
pi2 = estimar_pi2(a,b,c,d,e,n,pi)
```

Se aprecia en el código que la estimación 2 converge más rápido al valor de $\pi$ que la estimación 1.

<div class="alert alert-block alert-info"> <b> Problema 4. Eliminación Gaussiana - matriz de Hilbert. </b> </div>

Programe el método de eliminación Gaussiana sin pivoteo para resolver el sistema lineal $ Ax = b$.  Luego adapte el programa u observe como calcular $A^{-1}$. 
Testee el programa calculando la inversa de la matriz de Hilbert $H_9$. La matriz de Hilbert $H_n\in \mathbb R^{n\times n}$ se define  por
$$
(H_{n})_{i,j} = \frac{1}{i+j-1}
$$
para $i,j=1,...,n$. 

<div class="alert alert-block alert-success"> <b> Solución. </b> </div>

```python
import numpy as np

def create_Hn(n: int): 
    matrix = []
    for i in range(n):
        fila = []
        for j in range(n):
            fila.append((1/ ((i+1)+(j+1)-1)))
        matrix.append(fila)

    hilbert = np.array(matrix)
    return hilbert

def gauss_elim_sin_pivoteo(matriz: np.array, n: int, b_input):
    b = b_input.copy()

    # Eliminación progresiva
    for i in range(n):
        pivote = matriz[i][i]
        for k in range (i+1,n):
            r = matriz[k][i] / pivote
            matriz[k, i:] -= r * matriz[i, i:]
            b[k] -= r * b[i]
    
    # Sustitución regresiva
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        suma = np.dot(matriz[i, i+1:], x[i+1:])
        x[i] = (b[i] - suma) / matriz[i, i]

    return x

h9 = create_Hn(9)
input = np.random.rand(9)
solucion = gauss_elim_sin_pivoteo(h9, 9, input)
print(' Matriz Hilbert 9x9: ')
for fila in h9:
    print(fila)
print(f'\n Solución x = {solucion}')
```

Ahora adaptamos el código anterior para calcular la inversa, basta que los cambios en la sustitución progresiva y regresiva se realicen en la matriz identidad y no en un vector b.

```python
# Adaptación del programa para calcular A^-1

def inversa_A(matrix: np.array, n: int):

    matriz = matrix.copy()
    identidad = np.eye(n)

    # Eliminación progresiva
    for i in range(n):
        pivote = matriz[i][i]
        matriz[i] = matriz[i] / pivote #Dejamos la diagonal = 1
        identidad[i] = identidad[i] / pivote
        
        for k in range (i+1,n):
            r = matriz[k][i] 
            matriz[k, i:] -= r * matriz[i, i:]
            identidad[k, :] -= r * identidad[i, :]
    
    # Sustitución regresiva
    for i in range(n - 1, -1, -1):
        for k in range(i - 1, -1, -1):
            r = matriz[k, i]
            matriz[k, i:] -= r * matriz[i, i:]
            identidad[k, :] -= r * identidad[i, :]

    return identidad

h9 = create_Hn(9)
inversa = inversa_A(h9, 9)
print(' Inversa de A: ')
for fila in inversa:
    print(fila)
```
