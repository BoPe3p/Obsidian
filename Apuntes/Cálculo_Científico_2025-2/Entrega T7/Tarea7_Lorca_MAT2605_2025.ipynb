<div class="alert alert-block alert-info"> <b> Problema 1. Estimacion de error de regla de cuadratura  </b> </div>

Derive la cota del error para la regla de cuadratura de Simpson. 

<div class="alert alert-block alert-success"> <b> Solución. </b> </div>

La regla de Simpson aproxima la integral de una función f de la siguiente forma: 
$$
Q(f) = \frac{h}{3}[f(x_0)+4f(x_1)+f(x_2)] \ , \ x_0 = a \ , x_1 = \frac{a+b}{2} \ , \ h = \frac{b-a}{2}
$$

Además la Regla de Simpson sabemos que tiene precisión $= 3$ por lo que el error de la aproximación si es que se hace con un polinomio de grado 3 debería ser 0. Para derivar la cota del error expandimos la serie de Taylor de $f(x)$ alrededor de $x_1$. Trasladamos la serie al intervalo $[-h,h]$ para que el punto medio sea en 0.

Sea $g(t) = f(x_1+t)$, la expansión de Taylor queda:
$$
g(t) = g(0) + g'(0)t + \frac{g''(0)t^2}{2} + \frac{g^{(3)}(0)t^3}{6} + \frac{g^{(4)}(0)t^4}{24} + R_5(t)
$$

Calculamos la integral exacta, notemos que en un intervalo simétrico las derivadas impares se anulan:
$$
I=\int\limits_{-h}^h g(t)dt=  [g(0)t + \frac{g''(0)t^3}{6} + \frac{g^{(4)}(0)t^5}{120}]\big|_{-h}^h
$$
$$
I = [g(0)h + \frac{g''(0)h^3}{6} + \frac{h^{(4)}(0)h^5}{120}] - [-g(0)h + -\frac{g''(0)h^3}{6} + -\frac{h^{(4)}(0)h^5}{120}]
$$
$$
I = 2hg(0) + \frac{g''(0)h^3}{3} + \frac{h^{(4)}(0)h^5}{60}
$$

Evaluando $Q(g)$ en h y -h:
- $g(h) = g(0)+ hg'(0) + \frac{h^2}{2}​g''(0)+\frac{h^3}{6}​g'''(0)+\frac{h^4}{24}​g^{(4)}(0)+O(h^5)$
- $g(-h) = g(0)- hg'(0) + \frac{h^2}{2}​g''(0)-\frac{h^3}{6}​g'''(0)+\frac{h^4}{24}​g^{(4)}(0)+O(h^5)$

Luego en la regla de Simpson:
$$
Q = \frac{h}{3}[(g(-h) +g(h) +4g(0))] = \frac{h}{3}[(2g(0) + h^2g''(0) + \frac{h^4}{12}g^{(4)}(0) +4g(0))]
$$
$$
Q = 2hg(0) +\frac{h^3}{3}g''(0) + \frac{h^5}{36}g^{(4)}(0)
$$

Finalmente el error es la resta de I con Q:
$$
E = 2hg(0) + \frac{g''(0)h^3}{3} + \frac{h^{(4)}(0)h^5}{60} - 2hg(0) +\frac{h^3}{3}g''(0) + \frac{h^5}{36}g^{(4)}(0) = -\frac{1}{90}h^5g^{(4)}(0)
$$
Reemplazando los valores obtenemos $E = -\frac{1}{90}(\frac{b-a}{2})^5f^{(4)}(\zeta)$. Que es la cota que buscábamos.

<div class="alert alert-block alert-info"> <b> Problema 2. Implementación de reglas de cuadratura  </b> </div>

Derive los nodos y pesos de las siguientes reglas de cuadratura.
1. Cuadratura de Newton-Cotes de $5$ puntos equiespaciados.
2. Cuadratura Gaussiana con 2  puntos cerrada.

Considere las siguientes integrales

i. $$\int_{-1}^{1} \sin(3x) dx $$
ii.  $$\int_{1}^{3} \ln(x) dx $$
iii. $$ \int_{1}^{2} e^{x^{2}} dx $$
iv. $$\int_{-1}^{0} \cos(5x)\,dx $$
v. $$\int_{0}^{2} \exp(x^{-2}) \,dx $$
Aproxime las integrales utilizando las reglas de cuadratura anteriores.

<div class="alert alert-block alert-success"> <b> Solución. </b> </div>

Las fórmulas de cuadratura aproximan integrales definidas de la forma:
$$
\int\limits_a^b f(x)dx \approx \sum\limits_0^n A_jf(x_j)
$$

1. Newton Cotes cerrada:
$$
\sum\limits_0^n A_jf(x_j) \ : \ \text{(nodos:)} x_j = a+jh\ ,\ h=\frac{b-a}{n} \quad \text{(pesos:)} A_j = \int\limits_a^b l_{n,j}(x)dx
$$

Los nodos equiespaciados entonces son:
$$
x_0=a, \quad x_1=a+h. \quad x_2 = a+2h, \quad x_3 =a+3h, \quad x_4 = a +4h
$$

Los pesos se calculan haciendo (hagámoslo para $A_0$):
$$
A_0 = \int\limits_a^b l_{4,0}(x)dx \ , \ l_{4,0}(x) = \frac{(x-x_1)(x-x_2)(x-x_3)(x-x_4)}{(x_0-x_1)(x_0-x_2)(x_0-x_3)(x_0-x_4)}
$$
Reemplazando en los nodos:
$$
l_{4,0}(x) = \frac{(x-(a+h))(x-(a+2h))(x-(a+3h))(x-(a+4h))}{(-h)(-2h)(-3h)(-4h)}
$$
Con un cambio de variable para deshacernos de a, sea $x = a+th \ , \ t \in [0,4]\ dx = hdt$:
$$
l_{4,0}(x) = \frac{(th-h)(th-2h)(th-3h)(th-4h)}{(-h)(-2h)(-3h)(-4h)} = \frac{(t-1)(t-2)(t-3)(t-4)}{24}
$$
Finalmente tenemos que calcular:
$$
A_0 = h\int\limits_0^4 \frac{(t-1)(t-2)(t-3)(t-4)}{24} dt = \frac{7h}{45}
$$

Se procede de forma similar para el resto de pesos, la integral es casi la misma:
$$
A_1 = h\int\limits_0^4 \frac{t(t-2)(t-3)(t-4)}{-6} dt = \frac{32h}{45}
$$
$$
A_2 = h\int\limits_0^4 \frac{t(t-1)(t-3)(t-4)}{4} dt = \frac{12h}{45}
$$
$$
A_3 = h\int\limits_0^4 \frac{t(t-1)(t-2)(t-4)}{-6} dt = \frac{32h}{45}
$$
$$
A_4 = h\int\limits_0^4 \frac{t(t-1)(t-2)(t-3)}{24} dt = \frac{7h}{45}
$$

Finalmente la cuadratura queda:
$$
Q(f(x)) = \frac{h}{45}[7f(x_0) + 32f(x_1) + 12f(x_2) + 32f(x_3) + 2f(x_4)] 
$$

2. Cuadratura de Gauss:

En la última clase vimos que podemos calcular los nodos en el intervalo $[a,b]$, sabiendo que los nodos en $[-1,1]$ son los polinomios de Legendre, para $n$ puntos, calculamos las raices del polinomio $n+1$ y luego trasladamos los nodos a $[a,b]$ con la transformación: $x = (t+1)(\frac{b-a}{2})+a$. Los pesos se calculan: $A_j = \int\limits^3_1 \ell_{2,j}(x) dx $. También podemos ocupar el hecho de que la cuadratura de Gauss es exacta para calcular polinomios de grado k, luego también se puede calcular $\int x^k = \sum A_j x_j^k = \frac{x^{k+1}}{k+1}$. Obtenemos un sistema de ecuaciones para obtener los pesos. Igualmente tenemos los polinomios de Legendre, pesos y nodos tabulados:

Referencia: https://es.wikipedia.org/wiki/Cuadratura_de_Gauss

Para calcular los pesos tenemos la siguiente expresión:
$$
A_j = \frac{2}{(1-x_j^2)(P'_n(x_i))^2} \ , \ P_n = \text{ polinomio de Legendre de grado n}
$$

La misma página nos entrega los nodos y pesos para $n=2$ puntos:
$$
\text{Nodos:} \ \pm\sqrt{1/3} \quad \text{Pesos:} \ 1
$$

Igualmente lo anterior vale para el intervalo [-1,1], para cambiar los límites de integración se tiene que realizar el siguiente cambio de variable:
$$
\int\limits_a^b f(x) dx = \frac{b-a}{2} \int\limits_{-1}^1 f(\frac{b-a}{2} u + \frac{a+b}{2}) du \approx \frac{b-a}{2}\sum\limits_{i=1}^n A_i f(\frac{b-a}{2}u_i+\frac{a+b}{2})
$$

```python
# Aproximemos las integrales

def newton_cotes(f, a, b):
    n = 4
    h = (b-a)/n
    
    # Nodos equiespaciados
    nodos = np.array([a, a+h, a+2*h, a+3*h, b])
    pesos = np.array([7, 32, 12, 32, 7]) * h / 45

    return np.sum(pesos * f(nodos))

def gaussiana(f, a, b):
    # Para 2 puntos:
    nodos = np.array([-1/np.sqrt(3), 1/np.sqrt(3)])
    pesos = np.array([1.0, 1.0])

    if a == -1 and b == 1:
        return np.sum(pesos * f(nodos))

    def cambio_variable(u):
        return (b-a)/2*u + (a+b)/2
    
    resultado = 0
    for i in range(2):
        cambio = cambio_variable(nodos[i])
        resultado += pesos[i] * f(cambio)
    
    return resultado * (b-a)/2
```

```python
# Aproximemos las integrales:

nc1 = newton_cotes(lambda x: np.sin(3*x), -1, 1)
g1 = gaussiana(lambda x: np.sin(3*x), -1, 1)
print('Integral 1.\n')
print(f'Newton cotes: {nc1}')
print(f'Gaussiana: {g1}')
print('____________________________')


nc2 = newton_cotes(lambda x: x*np.log(x) - x, 1, 3)
g2 = gaussiana(lambda x: x*np.log(x) - x, 1, 3)
print('Integral 2.\n')
print(f'Newton cotes: {nc2}')
print(f'Gaussiana: {g2}')
print('____________________________')


nc3 = newton_cotes(lambda x: np.exp(x**2), 1, 2)
g3 = gaussiana(lambda x: np.exp(x**2), 1, 2)
print('Integral 3.\n')
print(f'Newton cotes: {nc3}')
print(f'Gaussiana: {g3}')
print('____________________________')


nc4 = newton_cotes(lambda x: np.cos(5*x), -1, 0)
g4 = gaussiana(lambda x: np.cos(5*x), -1, 0)
print('Integral 4.\n')
print(f'Newton cotes: {nc4}')
print(f'Gaussiana: {g4}')
print('____________________________')


nc5 = newton_cotes(lambda x: np.exp(x**(-2)), 0, 2)
g5 = gaussiana(lambda x: np.exp(x**(-2)), 0, 2)
print('Integral 5.\n')
print(f'Newton cotes: {nc5}')
print(f'Gaussiana: {g5}')
print('____________________________')
```

<div class="alert alert-block alert-info"> <b> Problema 3. Polinomio de aproximación </b> </div>

Determinar el grado de precision de las formulas para $\int_{-1}^1 f(x)dx$:
1. $Q_1(f) = \frac{4}{3} f(-0.5)-\frac{2}{3} f(0)+\frac{4}{3} f(0.5)$
2. $Q_2(f) = \frac{1}{4} f(-1)+\frac{3}{4} f(-1/3)+\frac{3}{4} f(1/3)+\frac{1}{4} f(1)$

<div class="alert alert-block alert-success"> <b> Solución. </b> </div>

Probemos con monomios de distintos grados:

- $n = 0$:
$$
\int_{-1}^1 1dx = 2
$$
$$
Q_1(1) = \frac{4}{3} 1-\frac{2}{3} 1+\frac{4}{3} 1 = 2 
$$

Ambas coinciden, seguimos aumentando el n hasta que no coincidan.

- $n = 1$:
$$
\int_{-1}^1 xdx = \frac{x^2}{2}\Big|^1_{-1} = 0
$$
$$
Q_1(x) = \frac{4}{3} (-0.5)-\frac{2}{3} (0) +\frac{4}{3} (0.5) = 0
$$

- $n = 2$:
$$
\int_{-1}^1 x^2 dx = \frac{x^3}{3}\Big|^1_{-1} = \frac{2}{3}
$$
$$
Q_1(x^2) = \frac{4}{3} (-0.5)^2-\frac{2}{3} (0)^2 +\frac{4}{3} (0.5)^2 = \frac{1}{3} + \frac{1}{3} = \frac{2}{3}
$$

- $n = 3$:
$$
\int_{-1}^1 x^3 dx = \frac{x^4}{4}\Big|^1_{-1} = 0
$$
$$
Q_1(x^3) = \frac{4}{3} (-0.5)^3-\frac{2}{3} (0)^3 +\frac{4}{3} (0.5)^3 = \frac{-0.5}{3} + \frac{0.5}{3} = 0
$$

- $n = 4$:
$$
\int_{-1}^1 x^4 dx = \frac{x^5}{5}\Big|^1_{-1} = \frac{2}{5}
$$
$$
Q_1(x^4) = \frac{4}{3} (-0.5)^4-\frac{2}{3} (0)^4 +\frac{4}{3} (0.5)^4 = \frac{0.25}{3} + \frac{0.25}{3} = \frac{1}{6}
$$

Grado de precisión de $Q_1 = 3$

Implementé  este código en python para hacerlo más rápido:

```python
# Integral exacta de x^n:
def integrar(grado, a=-1, b=1):
    return (b**(grado +1) - a**(grado +1)) / (grado +1)

# Ahora para determinar la precisiónd e los polinomios, definimos Q1 y Q2
def poli_q1(f, puntos=np.array([-0.5, 0, 0.5])):
    pesos = np.array([4/3, -2/3, 4/3])
    return np.dot(pesos, f(puntos))

def poli_q2(f, puntos=np.array([-1, -1/3, 1/3, 1])):
    pesos = np.array([1/4, 3/4, 3/4, 1/4])
    return np.dot(pesos, f(puntos))

print('Grado de precisión:\n')

max_precision_Q1 = -1
max_precision_Q2 = -1
limit_Q1 = True
limit_Q2 = True

for grad in range(0,6):
    f = lambda x: x**grad

    exact = integrar(grad)

    # Aproximaciones con los Q
    Q1 = poli_q1(f)
    Q2 = poli_q2(f)

    error_Q1 = abs(exact - Q1)
    error_Q2 = abs(exact - Q2)
    
    tol = 1e-10

    print(f"\nGrado {grad}: f(x) = x^{grad}")
    print(f"Integral exacta: {exact}")

    if error_Q1 < tol and limit_Q1:
        max_precision_Q1 = grad 
    else:
        limit_Q1 = False
        
    if error_Q2 < tol and limit_Q2:
        max_precision_Q2 = grad 
    else:
        limit_Q2 = False

# Obtenemos el grado máximo con el cual aproximan los monomios de grado n
print(f"Grado de precisión Q1: {max_precision_Q1}")
print(f"Grado de precisión Q2: {max_precision_Q2}")
```

Luego la precisión para ambas fórmulas es de $n=3$.

<div class="alert alert-block alert-info"> <b> Problema 4.</b> </div>

Para $f : R^2 \to R$ una funcion continua y $D = [0, 1]\times[0, 1]$, queremos deviar un formula para aproximar 
$$\int_{D} f(x,y) dx dy$$

1. Defina la funcion

$$
F(x) = \int_0^1 f(x, y) dy
$$ 

Entonces aproxime los valores $F(0), F( \frac{1}{2} ), F(1)$ con la regla de Simpson y luego aproxime $\int_0^1 F(x) dx$ usando otra vez la misma regla. Escriba la regla resultante.

1. Repita el procedimiento anterior para dar una formula de cuadratura sobre $D$ el triángulo de vértices $(0, 0), (0, 1), (1, 0)$. Pruebe que la fórmula anterior es exacta para $f(x, y) = x^2 + y^2$.

2. Implemente la regla de cuadratura sobre el triangulo y aproxime la integral de $f(x,y) = \sin(\pi(x-2y))$

<div class="alert alert-block alert-success"> <b> Solución. </b> </div>

Aproximemos los valores que nos piden, aplicando Simpson para la integral con respecto a y tenemos:
$$
F(x) = \int\limits_0^1 f(x,y)dy \approx \frac{b-a}{6}[f(a)+4f(\frac{a+b}{2})+f(b)] = \frac{1}{6}[f(x,0)+4f(x,\frac{1}{2})+f(x,1)]
$$
Luego:
- $F(0)\approx \frac{1}{6}[f(0,0)+4f(0,\frac{1}{2})+f(0,1)]$
- $F(\frac{1}{2})\approx \frac{1}{6}[f(\frac{1}{2},0)+4f(\frac{1}{2},\frac{1}{2})+f(\frac{1}{2},1)]$
- $F(1)\approx \frac{1}{6}[f(1,0)+4f(1,\frac{1}{2})+f(1,1)]$

Ahora aplicamos Simpson en x para aproximar la segunda integral:
$$
\int\limits_0^1 F(x)dx \approx \frac{1}{6}[F(0)+4F(\frac{1}{2})+F(1)]
$$
Reemplazando lo que se obtuvo anteriormente:
$$
\int\limits_0^1 F(x)dx \approx \frac{1}{36}[(f(0,0)+4f(0,\frac{1}{2})+f(0,1))+4(f(\frac{1}{2},0)+4f(\frac{1}{2},\frac{1}{2})+f(\frac{1}{2},1))+(f(1,0)+4f(1,\frac{1}{2})+f(1,1))]
$$

1. El triángulo que nos entregan define la siguiente integral:
$$
\int_0^1\int\limits_0^{1-x} f(x,y) dydx
$$
Repitiendo el proceso anterior:

$$
F(x) = \int\limits_0^{1-x} f(x,y) dy \approx \frac{1-x}{6} [ f(x,0) + 4f(x, \frac{1-x}{2}) + f(x, 1-x)]
$$

- $F(0)\approx \frac{1}{6}[ f(0,0) + 4f(0,\frac{1}{2})+f(0,1)]$
- $F(\frac{1}{2})\approx \frac{1}{12}[ f(\frac{1}{2},0) + 4f(\frac{1}{2},\frac{1}{4})+f(\frac{1}{2}\frac{1}{2})]$
- $F(1)\approx 0$

La segunda integral, aproximando nuevamente queda:
$$
\int\limits_0^1 F(x)dx \approx \frac{1}{36}[(f(0,0) + 4f(0,\frac{1}{2})+f(0,1))+4(f(\frac{1}{2},0) + 4f(\frac{1}{2},\frac{1}{4})+f(\frac{1}{2}\frac{1}{2}))]
$$

Para que la fórmula sea exacta para $f(x,y) =x^2+y^2$, calculemos primero:
$$
\int_0^1\int\limits_0^{1-x} x^2+y^2 dydx = \frac{1}{6} = 0.16666666666666666
$$

Con la aproximación nos da un resultado de:
- $F(0)\approx \frac{1}{6}[ f(0,0) + 4f(0,\frac{1}{2})+f(0,1)] = \frac{1}{3}$
- $F(\frac{1}{2})\approx \frac{1}{12}[ f(\frac{1}{2},0) + 4f(\frac{1}{2},\frac{1}{4})+f(\frac{1}{2}\frac{1}{2})] = \frac{1}{6}$

$$
\int_0^1 F(x)dx \approx \frac{1}{6}[\frac{1}{3}+ \frac{4}{6}] = \frac{1}{6}
$$

Por lo que se verifica que sí es exacto.

```python
# Aplicación con función seno
def cuadratura_triangulo_simpson(f):

    # Puntos en x: 0, 0.5, 1
    x_points = [0, 0.5, 1]
    pesos_x = [1/6, 4/6, 1/6]
    
    resultado = 0
    
    for i, x in enumerate(x_points):
        # Para cada x, los límites en y son [0, 1-x]
        y_max = 1 - x

        if y_max == 0:
            F_x = 0
        else:
            # Puntos en y: 0, y_max/2, y_max
            y_points = [0, y_max/2, y_max]
            pesos_y = [1/6, 4/6, 1/6]
            
            # Integral con y
            F_x = 0
            for j, y in enumerate(y_points):
                F_x += pesos_y[j] * f(x, y) * y_max 
        
        resultado += pesos_x[i] * F_x
    
    return resultado

f = lambda x, y: np.sin(np.pi*(x-2*y))
resultado = cuadratura_triangulo_simpson(f)
print('=== REGLA DE CUADRATURA APLICADA SOBRE EL TRIÁNGULO PARA APROXIMAR SENO ===')
print(f' Resultado: {resultado}')
```

Con este resultado, concluimos dado la aproximación que la integral sobre el triángulo de $f(x) = \sin(\pi(x-2y)) = 0$.
